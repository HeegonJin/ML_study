```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(w0, w1):
    return w0 ** 2 + 2 * w0 * w1 + 3

def df_dw0(w0, w1):
    return 2 * w0 + 2 * w1

def df_dw1(w0, w1):
    return 2 + w0
```

```python
w0 = np.arange(-2, 2, 0.25)
w1 = np.arange(-2, 2, 0.25)
wn = len(w0)

ww0, ww1 = np.meshgrid(w0, w1)
ff = np.zeros((len(w0), len(w1)))
dff_dw0 = np.zeros_like(ff)
dff_dw1 = np.zeros_like(ff)
for i0 in range(len(ww0)):
    for i1 in range(len(ww1)):
        ff[i1, i0] = f(w0[i0], w1[i1])
        dff_dw0[i1, i0] = df_dw0(w0[i0], w1[i1])
        dff_dw1[i1, i0] = df_dw1(w0[i0], w1[i1])
```

```python
fig, ax = plt.subplots(1, 2)

plt.rcParams["figure.figsize"] = (9,4)
plt.rcParams['figure.subplot.wspace'] = 0.3

cs = ax[0].contour(ww0, ww1, ff, 10, colors='k')
ax[0].clabel(cs, fmt='%2.0f', fontsize=8)
ax[0].set_xticks(range(-2, 3, 1))
ax[0].set_yticks(range(-2, 3, 1))
ax[0].set_xlim(-2.5, 2.5)
ax[0].set_ylim(-2.5, 2.5)
ax[0].set_xlabel('$w_0$', fontsize=14)
ax[0].set_ylabel('$w_1$', fontsize=14)

ax[1].quiver(ww0, ww1, dff_dw0, dff_dw1)
ax[1].set_xticks(range(-2, 3, 1))
ax[1].set_yticks(range(-2, 3, 1))
ax[1].set_xlim(-2.5, 2.5)
ax[1].set_ylim(-2.5, 2.5)
ax[1].set_xlabel('$w_0$', fontsize=14)
ax[1].set_ylabel('$w_1$', fontsize=14)
plt.show()
```

```python
def softmax(x0, x1, x2):
    u = np.exp(x0) + np.exp(x1) + np.exp(x2)
    return np.exp(x0) / u, np.exp(x1) / u, np.exp(x2) / u
```

```python
y = softmax(2, 1, 1)
```

```python
from mpl_toolkits.mplot3d import Axes3D

xn = 20
x0 = np.linspace(-4, 4, xn)
x1 = np.linspace(-4, 4, xn)
y = np.zeros((xn, xn, 3))

for i0 in range(xn):
    for i1 in range(xn):
        y[i1, i0, :] = softmax(x0[i0], x1[i1], 1)

xx0, xx1 = np.meshgrid(x0, x1)
plt.figure(figsize=(8,3))
for i in range(2):
    ax = plt.subplot(1, 2, i+1, projection='3d')
    ax.plot_surface(xx0, xx1, y[:,:, i], 
                   rstride=1, cstride=1, alpha=0.3,
                    color='blue', edgecolor='black')
    ax.set_xlabel('$x_0$', fontsize=14)
    ax.set_ylabel('$x_1$', fontsize=14)
    ax.view_init(40, -125)


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
%matplotlib inline
```

```python
def gauss(x, mu, sigma):
    N, D = x.shape
    c1 = 1 / (2 * np.pi) ** (D / 2)
    c2 = 1 / (np.linalg.det(sigma)**(1/2))
    inv_sigma = np.linalg.inv(sigma)
    c3 = (x - mu)
    c4 = np.dot(c3, inv_sigma)
    c5 = np.zeros(N)
    for d in range(D):
        c5 = c5 + c4[:, d] * c3[:, d]
    p = c1 * c2 * np.exp(-c5 / 2)
    return p
```

```python
x = np.array([[1, 2], [2, 1], [3, 4]])
mu = np.array([1, 2])
sigma = np.array([[1, 0], [0, 1]])
print(gauss(x, mu, sigma))
```

```python
X_range0 = [-3, 3]
X_range1 = [-3, 3]

def show_contour_gauss(mu, sig):
    xn = 40
    x0 = np.linspace(X_range0[0], X_range0[1], xn)
    x1 = np.linspace(X_range1[0], X_range1[1], xn)
    xx0, xx1 = np.meshgrid(x0, x1)
    x = np.c_[np.reshape(xx0, (xn * xn)), np.reshape(xx1, (xn * xn))]
    f = gauss(x, mu, sig)
    f = f.reshape(xn, xn)
    cont = plt.contour(xx0, xx1, f, 15, colors='k')
    plt.grid(True)
    

def show3d_gauss(ax, mu, sig):
    xn = 40
    x0 = np.linspace(X_range0[0], X_range0[1], xn)
    x1 = np.linspace(X_range1[0], X_range1[1], xn)
    xx0, xx1 = np.meshgrid(x0, x1)
    x = np.c_[np.reshape(xx0, (xn * xn)), np.reshape(xx1, (xn * xn))]
    f = gauss(x, mu, sig)
    f = f.reshape(xn, xn)
    ax.plot_surface(xx0, xx1, f, rstride=2, cstride=2, alpha=0.3, color='blue', edgecolor='black')
    

mu = np.array([1, 0.5])
sigma = np.array([[2, 1], [1, 1]])
Fig = plt.figure(1, figsize=(7, 3))
Fig.add_subplot(1,2,1)
show_contour_gauss(mu, sigma)
plt.xlim(X_range0)
plt.ylim(X_range1)
plt.xlabel('$x_0$', fontsize=14)
plt.ylabel('$x_1$', fontsize=14)
Ax = Fig.add_subplot(1,2,2,projection='3d')
show3d_gauss(Ax, mu, sigma)
Ax.set_zticks([0.05, 0.10])
Ax.set_xlabel('$x_0$', fontsize=14)
Ax.set_ylabel('$x_1$', fontsize=14)
Ax.view_init(40, -100)
plt.show()
```
